```python
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
```

### Нормальное распределение

```python
# задаём мат ожидание и отклонение среднеквадратическое
mu = np.array([2, 3, 1, 2, 2])
sigma = np.array([0.5, 0.5, 0.5, 1.5, 0.1])
```

```python
legend_1 = np.array([])
for i in range(0, len(mu)):
    norm_rv = sts.norm(loc=mu[i], scale=sigma[i])
    x = np.linspace(0, 4, 100)
    cdf = norm_rv.cdf(x)
    legend_1 = np.append(legend_1, ['mu={} sigma={}'.format(mu[i], sigma[i])])
    plt.plot(x, cdf)
plt.legend(legend_1, loc='upper right', bbox_to_anchor=(1.45, 1.0))   
plt.grid()
plt.title('Вероятностные распределения (Гауссово распределения)')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.show()
```

```python
legend_2 = np.array([])
for i in range(0, len(mu)):
    norm_rv = sts.norm(loc=mu[i], scale=sigma[i])
    x = np.linspace(0, 4, 100)
    pdf = norm_rv.pdf(x)
    legend_1 = np.append(legend_2, ['mu={} sigma={}'.format(mu[i], sigma[i])])
    plt.plot(x, pdf)
plt.legend(legend_2, loc='upper right', bbox_to_anchor=(1.45, 1.0))   
plt.grid()
plt.title('Плотность вероятности (Гауссово распределения)')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

### Экспоненциальное распределение

```python
# задаём параметр лямда 
lam = np.array([0.1, 0.001, 0.5, 1.5, 2])
```

```python
legend_3 = np.array([])
for i in range(0, len(lam)):
    expon_rv = sts.expon(lam[i])
    x = np.linspace(0, 4, 100)
    cdf = expon_rv.cdf(x)
    legend_3 = np.append(legend_3, ['lam={}'.format(lam[i])])
    plt.plot(x, cdf)
plt.legend(legend_3, loc='upper right', bbox_to_anchor=(1.45, 1.0))   
plt.grid()
plt.title('Вероятностные распределения (Экспоненциальное распределение)')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.show()
```

```python
legend_4 = np.array([])
for i in range(0, len(lam)):
    expon_rv = sts.expon(lam[i])
    x = np.linspace(0, 4, 100)
    pdf = expon_rv.pdf(x)
    legend_4 = np.append(legend_4, ['lam={}'.format(lam[i])])
    plt.plot(x , pdf)
plt.legend(legend_4, loc='upper right', bbox_to_anchor=(1.45, 1.0))   
plt.grid()
plt.title('Плотность вероятности (Экспоненциальное распределение)')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

### Стьюдента

```python
# задаём степени свододы для распределения Стьюдента
n = np.array([10, 0.01, 0.2, 0.8, 2])
```

### Задание 2
Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания в отрезок $[3.5;7]$, т.е. P(3.5 < $\xi$ < 7).

```python
# у нас С - константа, поэтому равномерное распределение, а это значит, что С = 1 / (b - a)
a = 0
b = 5
```

```python
C = 1 / (b - a) 
print('Константа равна ', C)
```

```python
# найдём мат ожидание, поскольку равномерное распределение, то это просто середина
mu = (a + b) / 2 
print('Матожидание', mu)
```

```python
# вероятность попадания в отрезок   [3.5;7] , найдём через площадь плотности вероятности от 3,5 до 5, т.к больше 5 уже ноль
a1 = 3.5
P = C * (b - a1)
print('P(3.5<𝜉<7) = % .2f' % P)
```

### Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
import math
```

```python
# первая ситуация
t_red = 54
t_yellow = 4
t_green = 28
```

```python
time_1_circle = t_red + t_yellow + t_green
time_1_circle
```

```python
p_red = t_red / time_1_circle
p_yellow = t_yellow / time_1_circle
p_green = t_green / time_1_circle
```

```python
I =  - (p_red * math.log2(p_red) + p_yellow * math.log2(p_yellow) + p_green * math.log2(p_green))
print('Количество информации о цвете %.3f бит' % I)
```

```python
# вторая ситуация 
t = 28
```

```python
time_1_circle = t * 3
time_1_circle
```

```python
p = t / time_1_circle
p 
```

```python
I2 = math.log2(1 / p)
print('Количество информации о цвете %.3f бит' % I2)
```
