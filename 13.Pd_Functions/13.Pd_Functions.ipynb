### Задание 1

**Домашнее задание к лекции.

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

### Решение

```python
import pandas as pd

def class_movies(rat):
    """Функция классифицирует рейтинги на низкий(<2), средний(>2,<=4) и высокий(>4)"""
    if rat <= 2:
        return 'low rating'
    
    elif (rat <= 4) & (rat > 2):
        return 'middle rating'
    
    elif (rat <= 5) & (rat > 4):
        return 'high rating'

table_data = pd.read_csv('ratings.csv')
table_data['class'] = table_data['rating'].apply(class_movies)

table_data.groupby('class').head()
```

### Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

### Решение

```python
def geo_class(reque):
    """Функция определяет по вхождению города в запрос(keyword) регион"""
    geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    #разделение запроса на слова через пробел в список
    list_reque = reque.split(' ')
    #поиск в запросе вхождение в отобранный словарь геоданных
    for key, towns in geo_data.items():
        for num_town in towns:
            if num_town in list_reque:
                out = key
                return out
            else:
                out = 'undefined'
    return out

table_data2 = pd.read_csv('keywords.csv')
table_data2['region'] = table_data2['keyword'].apply(geo_class)
table_data2.groupby('region').head()
```

### Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

P/S/ все файлы находятся в архиве «дополнительные файлы к лекции и дз»

### Решение

```python
import re
# заполнение списка годами 1950-2010 (2011 не входит в range)
years = []
for year in range(1950, 2011, 1):
    years.append(year)

def production_year(name_movie):
    """Функция выделяет год фильма или ставит 1900, если его нет при помощи библиотеки регулярных выражений"""
    if re.findall(r'\d{4}', name_movie):
        year = re.findall(r'\d{4}', name_movie)
        return year[0]
    else:
        return 1900
# чтение файла с фильмами и фильтрация по нужным столбцам
data_movies = pd.read_csv('movies.csv')
data_movies['year'] = data_movies['title'].apply(production_year)
data_movies = data_movies[['movieId', 'title', 'year']]
# чтение файла с рейтингом и объединение двух датафреймов при помощи метода merge и по типу: outer чтобы все столбцы обеих были
data_ratings = pd.read_csv('ratings.csv')
general_data = data_movies.merge(data_ratings, how = 'outer')
# выделяем в новой таблице нужные нам столбцыЮ группируем по году, считаем среднее по рейтингу и сортируем по убыванию
general_data = general_data[['year', 'rating']].groupby(['year']).mean().sort_values(by=['rating'], ascending=False)
general_data.head(20)

```
