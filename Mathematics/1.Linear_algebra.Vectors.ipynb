# Домашнее задание "Линейная алгебра. Векторы".¶
### Даны вектора x = (1,1), y = (2, 0) и z = (0,2)

### Надо:

- Изобразить вектора на экране
- Изобразить точку x + y + z
- Найти угол между векторами x и y

### Решение

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,1])
y = np.array([2,0])
z = np.array([0,2])
```

### Изобразить вектора на экране

```python
plt.figure()
plt.plot(x[0], x[1], 's', color='r')
plt.plot(y[0], y[1], 'p', color='b')
plt.plot(z[0], z[1], '*', color='g')
plt.grid()
```

```python
#plt.quiver для построения стрелочки
# scale_units='xy' - масштабирование стрелок в зависимости от значений по осям
# angles='xy' - определение угла стрелок
plt.quiver([0, 0, 0], [0, 0, 0], [x[0], y[0], z[0]], [x[1], y[1], z[1]], angles='xy', scale_units='xy', color=['r','b','g'], scale=1)
plt.xlim(-0.1, 2.1)
plt.ylim(-0.1, 2.1)
plt.grid()
```

### Изобразить точку x + y + z

```python
v = x+y+z
v
```

```python
plt.figure()
plt.plot(x[0], x[1], 's', color='r')
plt.plot(y[0], y[1], 'p', color='b')
plt.plot(z[0], z[1], '*', color='g')
plt.plot(v[0], v[1], 'o', color='k')
plt.grid()
```

```python
plt.quiver([0, 0, 0, 0], [0, 0, 0, 0], [x[0], y[0], z[0], v[0]], [x[1], y[1], z[1], v[1]], angles='xy', scale_units='xy', color=['r','b','g', 'k'], scale=1)
plt.xlim(-0.1, 3.1)
plt.ylim(-0.1, 3.1)
plt.grid()
```

### Найти угол между векторами x и y

```python
alpha = np.rad2deg(np.arccos(np.dot(x,y) / (np.linalg.norm(x) * np.linalg.norm(y))))
alpha
```
