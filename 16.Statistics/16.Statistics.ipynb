# Задание 2. Работа с выбросами

```python
# найдём выбросы в каждом столбце
dict_border_outliers = {}
for column in df:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_out = Q1 - (1.5 * IQR)
#так как здесь не может быть отрицательной границы, то ставим минимальное значение, обозначающие наши пропуски (!просто нули здесь невозможны)
    if lower_out < 0: lower_out = 0
    higher_out = Q3 + (1.5 * IQR)
    dict_border_outliers.setdefault(column, [lower_out, higher_out])
dict_border_outliers
```

```python
# выведем на экран отдельно исследуемые столбцы без выбросов, очерчивая границу посчитанную выше
# проанализируем ситуацию
for column in df:
    outliers = df[df[column].between(dict_border_outliers[column][0], dict_border_outliers[column][1],inclusive=True)].sort_values(column)
    print(column,'\n', outliers[column],'\n')
```

```python
# выведем на экран отдельно исследуемые столбцы без выбросов, очерчивая границу посчитанную выше
# проанализируем ситуацию
for column in df:
    df = df[df[column].between(dict_border_outliers[column][0], dict_border_outliers[column][1], inclusive=True)]
df
```

## Выбросы и их причины 
У нас всего 299 строк значений, разберём те, у кого отсеились выбросы. Итого, осталось строк у следующих столбцов: 
Surgery - 298; Age - 275, Hospital Number - 263, rectal temperature - 225, pulse - 270, respiratory rate - 224; temperature of extremities - 243; peripheral pulse - 230

Причины:

Age - скорее ошибки ввода данных(человеческий фактор), либо ошибки обработки данных -попадение случайно, т.к. к примеру число 9 не обозначает ничего в описании столбца;

Hospital Number - здесь так же ошибки ввода данных(человеческий фактор)/опечатка, случай введения на одну цифру больше, чем должно быть, однако в теории, возможно, всё-таки id лошадки, но слишком сильно выбивается номер на порядок больше остальных, поэтому я бы не учитывал эти данные или проконсультировался со сборщиком данных=); 

rectal temperature - из описания "этот параметр обычно изменяется по мере развития проблемы, например, может начинаться нормально, затем повышаться из-за поражения, переходя обратно в нормальный диапазон, когда лошадь впадает в шок" - трудно сказать, что здесь выброс, а что реальные значения; 

pulse - по описанию, нормальный пульс может быть у спортивных лошадей 20-25 (30-40 взрослые), а с болезненными поражениями или в шоке может быть повышенный пульс, сильных выбросов нет, которые есть, это только не совершенные рамки выбросов, однако есть и значение 180 скорее ошибка измерителя; 

respiratory rate - как написано в описании столбца, полезность сомнительна из-за больших колебаний, так что выход за рамки лишь условное ограничение, как и условны сами ограничения

Решение по выбросам

Можно использовать условно принятый интервал через IQR и Q1, Q2, исключив все выбросы, которые выходят за рамка интервала, остальные значения выходящие за рамки можно либо принять за погрешность нашего метода расчёта, либо расширять рамки до адекватных возможных значений

Итого, после чистки от строк с выбросами, уменьшилось количество данных на 60%, что не супер (было 298, стало 120) Зато они все чистенькие=)

```python
df = df.dropna(thresh=7)
df
```

```python
quality_data = ['surgery', 'Age', 'Hospital Number', 'temperature of extremities', 'peripheral pulse']
quantitative_data = ['rectal temperature', 'pulse', 'respiratory rate']
```

```python
dg = df
dg
```

```python
#распределение замен пропусков: количественный тип данных - медиана, качественный - мода
for col in dg.columns:
    if col in quality_data: 
        dg = dg[col].fillna(df.groupby(['surgery'])[col].mode()[0], inplace = True)
    elif col in quantitative_data: 
        dg = dg[col].fillna(df.groupby(['surgery'])[col].transform('median'), inplace = True)
dg
```
