## Уровень 0:
## Посчитайте значение первой и второй производной функции

### $f(x) = x^5 + 4sin(2x) + cos(3x+3)$
В точке x = 1.

```python
import numpy as np
from scipy.misc  import derivative 
from sympy import *
import matplotlib.pyplot as plt
```

```python
x = symbols('x')

f_x = x ** 5 + 4 * sin(2 * x) + cos(3 * x + 3)
print("Before Differentiation : f(x) = {}".format(f_x))

def f(x):
    return x ** 5 + 4 * sin(2 * x) + cos(3 * x + 3)

f1 = derivative(f, 1.0, dx=1e-6, n=0)

print("f'(x=1) = {}".format(f1))
```

```python
dif = diff(f_x, x)
  
print("After First Differentiation : f'(x) = {}".format(dif))

df = derivative(f, 1.0, dx=1e-6)

print("f'(x=1) = {}".format(df))
```

```python
difdif = diff(dif, x)
  
print("After Second Differentiation : f''(x) = {}".format(difdif))

ddf = derivative(f, 1.0, dx=1e-6, n=2)

def dif_dif_f(x):
    return difdif
print("f''(x=1) = {}".format(ddf))
```

## Уровень 1:

### Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].


```python
x = symbols('x')

f_x =  sin(2 * x) ** 5 
print("Before Differentiation : f(x) = {}".format(f_x))

def f(x):
    return sin(2 * x) ** 5 
```

```python
# создаём массив данных для переменной х
X = np.linspace(-5, 5)
```

```python
# считаем значение функции f(x) и забиваем в "массив" на отрезке [-5,5]
F = []
for i in range(0, len(X)):
    F.append(derivative(f, X[i], n=0))
```

```python
# строим график функции f(x)
plt.plot(X, F)
plt.xlabel("x = [-5, 5]")
plt.ylabel("f(x)")
plt.title('График f(x)')
plt.grid()
```

```python
dif = diff(f_x, x)
  
print("After First Differentiation : f'(x) = {}".format(dif))
```

```python
dF = []
for i in range(0, len(X)):
    dF.append(derivative(f, X[i], dx=1e-6, n=1))
```

```python
# строим график первой производной функции f(x)
plt.plot(X, dF)
plt.xlabel("x = [-5, 5]")
plt.ylabel("f'(x)")
plt.title('Первая производная f(x)')
plt.grid()
```

```python
difdif = diff(dif, x)
  
print("After Second Differentiation : f''(x) = {}".format(difdif))
```

```python
ddF = []
for i in range(0, len(X)):
    ddF.append(derivative(f, X[i], dx=1e-6, n=2))
```

```python
# строим график второй производной функции f(x)
plt.plot(X, ddF)
plt.xlabel("x = [-5, 5]")
plt.ylabel("f''(x)")
plt.title('Вторая производная f(x)')
plt.grid()
```
