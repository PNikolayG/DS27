```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('country_vaccinations.csv')
df
```

```python
df.info()
```

```python
# сделаем таблицу для суммарной вакцинации по странам за период за период 2020-12-13 - 2021-02-25
total_dynamics_by_country = df[['daily_vaccinations', 'date', 'country']].groupby('country').sum()
total_dynamics_by_country
```

```python
# сделаем таблицу для суммарной вакцинации по датам по всем странам
total_dynamics_by_date = df[['daily_vaccinations', 'date']].groupby('date').sum()
total_dynamics_by_date
```

```python
# динамика по вакцианции в странах
total_dynamics_by_country[total_dynamics_by_country['daily_vaccinations'] > 3000000].plot(kind='bar', colormap='Paired')
```

```python
total_dynamics_by_country[total_dynamics_by_country['daily_vaccinations'] > 3000000].plot(kind='barh', 
                                        title='Total vaccinations')
```

### Вывод 
Сразу видим на вертикальной и горизонтальной диаграммах лидеров по вакцинации: США, Китай, Англия

```python
# динамика вакцинации по дням по всему миру
# total_dynamics_by_date.plot(kind='bar')
sns.heatmap(total_dynamics_by_date)
```

### Вывод
Можно заметить на тепловой карте, что начиная с января достаточно активно начали вакцинировать весь мир, по крайней мере вышли за отметку  3 000 000 вакцин в день

```python
total_dynamics_by_date.sort_values(by = 'daily_vaccinations', ascending = False).head(3)
```

```python
top_for_pie = total_dynamics_by_date['daily_vaccinations'].sort_values(ascending = False).head(3)
minx = top_for_pie.plot(kind='pie', title='Top minimum dates')
```

```python
total_dynamics_by_date.sort_values(by = 'daily_vaccinations', ascending = True).head(4)
```

```python
top_for_pie = total_dynamics_by_date['daily_vaccinations'].sort_values(ascending = True).head(4)
minx = top_for_pie.plot(kind='pie', title='Top maximum dates')
```

### Вывод
Здесь мы можем убедиться в достаточно логичном заключении. Когда только начинали вакцинацию в первые три дня, было меньше всего вакцинируемых, а под конец уже в феврале 7,8,9 число были самыми вакцинируемыми датами в рамках мира
