### Решение

```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

# p  
def owner_from_number():
        """
        Данная функция определяет владельца документа - по номеру этого документа
        """
    #вводим номер документа строчного типа
        num = str(input('Введите номер документа: '))
    #создаём цикл, который проходится по элементам=словарям в списке
        for dictionary in documents:
        #проверяем на равенство введенного значения с значения по ключу словаря
            if num == dictionary['number']:
            #при этом создаём переенную на вывод, записывая в неё результат вывода: нашлось или нет значение в списке из словарей
            #как нашлось выход из цикла и вывод результата
                out = f"Владелец документа: {dictionary['name']}"
                break
            else:
                out = 'Документ не найден в базе'
        print(out)

# s
def shelf_from_number(num_s):
    """
    Данная функция определяет номер полки, на которой лежит документ - по номеру этого документа
    """
  #вводим номер документа строчного типа
#     num = str(input('Введите номер документа: '))
    #разделяем ключ и значение словаря по переменным
    for key, values in directories.items():
        #если введённый номер есть в значении словаря, выводим ключ и останавливаем цикл
        #вводим две переменные на вывод, одна ответственна за ответ-полку, вторая за описание результата
            out_answer = 0
            if num_s in values:
                out_print = 'Документ хранится на полке: '
                out_answer = key
                break
            else:
                out_print = 'Документ не найден в базе '
    return ([out_answer, out_print])

# I
#вывод значений словаря с использованием функции выяснения полки по номеру документа
def information():
    """
    Данная функция выводит информацию о всех документах в архиве
    """
    #пробегаем по словарям из списка, выводя все их значения по ключам при помощи f строки
    for numb in documents:
        print(f"№: {numb['number']} тип: {numb['type']}, владелец: {numb['name']}, полка хранения: {shelf_from_number(numb['number'])[0]}")


# ads
def add_shelves():
    """
    Данная функция добавляет уникальную полку в словарь
    """
    in_shelf = str(input('Введите номер полки: '))
    #проверка введённой полки в словаре полок по ключам
    if in_shelf in directories:
        #если есть, то вывод описания и результата через запятую методом join
            out_answer = list(directories.keys())
            print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(out_answer)}")
    else:
        #если нет, то добавления в словарь setdefault + вывод описания и результата через запятую методом join
            directories.setdefault(f'{in_shelf}', [])
            out_answer = list(directories.keys())
            print(f"Полка добавлена. Текущий перечень полок: {', '.join(out_answer)}")

# ds
def remove_empty_shelf():
    """
    Данная функция удаляет полку, проверив перед этим есть ли что-то на ней
    """
    #вводим переменную, которая равна введённой полке
    in_shelf = str(input('Введите номер полки: '))
    #проверяем есть ли введённая полка в словаре
    if in_shelf in directories:
        #првоеряем условие, в значении словаря пусто, если да - удаляем эту полку=значение словаря
        if directories[in_shelf] == []:
            del(directories[in_shelf])
            print(f"Полка удалена. Текущий перечень полок: {', '.join(directories)}")
            #если не пусто в значении по ключу, на полке есть какой-то документ, то печаем фразу
        elif directories[in_shelf] != []:
            print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {', '.join(directories)}")
    #а если введённой полки нет в словаре, то пишем фразу
    elif in_shelf not in directories:
        print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories)}")
        
# ad
def add_document():
    """
    Данная функция позволяет добавить новый документ
    """
    #создаём словарь=транспорт в список из словарей для введённых значений
    new_dictionary = {}
    #помещаем его в список слвоарей=данных о документе
    documents.append(new_dictionary)
    #вводим параметры нового документа
    a = str(input('Введите номер документа: '))
    b = str(input('Введите тип документа: ')) 
    c = str(input('Введите владельца документа: ')) 
    d = str(input('Введите полку для хранения: ')) 
    #проверка условия: есть ли полка в словаре
    if d in directories:
        #если есть, то пополняем по ключу значение=список словаря при помощи append()
        directories[d].append(a)
        #пополняем введённый и добавленный словарь введёнными параметрами документа
        new_dictionary.setdefault('number', a)
        new_dictionary.setdefault('type', b)
        new_dictionary.setdefault('name', c)
        print('Документ добавлен. Текущий список документов:')
    else:
        print('Такой полки не существует. Добавьте полку командой ads. Текущий список документов:')
    #в конце любого исхода выводим текущий спсиок документов при помощи написанной функции information()
    information()


# d проблемка с I функцией
def delete_document():
    """
    Данная функция позволяет удалить документ из данных
    """
    #вводим переменную для ввода номера документа
    in_doc = str(input('Введите номер документа: '))
    #цикл перебирающий элементы списка=словари
    for dictionar in documents:
        #проверка наличия введённого номера документа в значениях словаря 
        if in_doc in dictionar.values():
            #если есть, то узнаём номер полки по номеру документа при помощи написанной функции shelf_from_number(), берём 0-й вывод
            shelf = shelf_from_number(in_doc)[0]
            #удаляем по ключу от переменной shelf конкретный элемент=номер введённого документа в списке при помощи .remove()
            directories[f'{shelf}'].remove(in_doc)
            #далее удаляем информацию об этом документе в списке словарей при помощи del() в переменной перебора словарей dictionar
            del(dictionar['number'])
            del(dictionar['type'])
            del(dictionar['name'])
            #вводим переменную вывода текста, в случае находки элемента break выход из цикла
            out_print = 'Документ удален. Текущий список документов:'
            break
        else:
            out_print = 'Документ не найден в базе. Текущий список документов:'
    #вывод переменной теста и всей информации о документах
    print(out_print)
    information()
    
# m
def bear_document():
    """
    Данная функция позволяет переместить документ из полки на полку
    """
    #введём две переменные ввода данных номера документа и полки
    in_doc = str(input('Введите номер документа: ')) 
    info = shelf_from_number(in_doc)  
    #условие продолжение работы, есть ли документ или нет при помощи функции shelf_from_number()
    if info[1] == 'Документ не найден в базе ':
        out = 'Документ не найден в базе. Текущий список документов: '
        print(out)
        information()
    else:
        in_shelf = str(input('Введите номер полки: '))
        #цикл перебор по словарю x=ключ, y=значение
        for ke, val in directories.items():
                #условие если номер полки равен ключу, т.е. есть ли такая полка
                if in_shelf == ke:
                    #если да, то сначала находим полку, где лежал документ по shelf_from_number()
                    num_shelf = shelf_from_number(in_doc)[0]
                    #удаляем элемент списка in_doc в значении словаря по ключу num_shelf
                    directories[num_shelf].remove(in_doc)
                    #добавляем d введённую полку документ
                    directories[in_shelf].append(in_doc)
                    out = 'Документ перемещен. Текущий список документов: '
                    print(out)
                    information()
                    break
                else: 
                    out = f"Такой полки не существует. Текущий перечень полок: {', '.join(directories)}"                 
    if out == f"Такой полки не существует. Текущий перечень полок: {', '.join(directories)}":
        print(out) 
           
        

while True:
    user_push = str(input('Введите команду: ')) 
    if user_push == 'q':
        break
    elif user_push == 'p':
        owner_from_number()    
    elif user_push == 's':
        num_s = str(input('Введите номер документа: '))
        answer = shelf_from_number(num_s)
        if answer[0] == 0:
            print(answer[1])
        else:
            print(answer[1], answer[0])
    elif user_push == 'I':
        information()
    elif user_push == 'ads':
        add_shelves()
    elif user_push == 'ds':
        remove_empty_shelf()
    elif user_push == 'ad':
        add_document()
    elif user_push == 'd':
        delete_document()
    elif user_push == 'm':
        bear_document()   
```
