### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений


```python
 print(f'{123.444:.2f}')
```

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.tail()
```

```python
# создадим столбец дублер, начиная как бы с самого широкого условия - все остальные варианты без изменений
log['source_type'] = log['traffic_source']
log.head()
```

```python
# отбор по гуглу и яндексу, замена
log.loc[log.traffic_source.str.contains('yandex' and 'google'), 'source_type'] = 'organic'
log.head()
```

```python
# далее сделали для всех кто paid и email other
log.loc[log.traffic_source == 'paid', 'source_type'] = 'other'
log.loc[log.traffic_source == 'email', 'source_type'] = 'other'
log.tail()
```

```python
# далее сделали для всех кто paid и email other, а потом сузили круг, для тех кто из РФ заменить на ad
log.loc[(log.region == 'Russia') & (log.traffic_source == 'paid'), 'source_type'] = 'ad'
log.loc[(log.region == 'Russia') & (log.traffic_source == 'email'), 'source_type'] = 'ad'
log.tail()
```

### Задание 3

1. Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


```python
film_marks = pd.read_csv('ml-latest-small/ratings.csv')
film_marks.head()
```

```python
# группируем данные в одну таблицу для всех данных с индексами по user, для рейтинга считаем количество оценок, для timestamp
# считаем минимальное и максимальное значение и среднее время жизни? после отбор пользователей с более 100 оценками
processing = film_marks.groupby('userId').agg({'rating': 'count', 'timestamp': ['min', 'max']})
processing['mean_life'] = processing['timestamp']['max'] - processing['timestamp']['min']
final_film = processing.loc[processing['rating']['count'] > 100]
final_film
```

### Задание 4

2. Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия - ниже). Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента


Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

### Необходимо сформировать таблицу с тремя типами выручки для каждого client_id без указания адреса клиента

```python
# берё за основу таблицу с наибольшим количеством id, т.е. client_base - общую и подклеиваем остальные по параметру client_id
# только для начала фильтруем по одному столбцу, т.к. просили без адресов
revenue = client_base
revenue = revenue[['client_id']].merge(air, on = 'client_id', how='left')
revenue = revenue.merge(auto, on = 'client_id', how='left')
revenue = revenue.merge(rzd, on = 'client_id', how='left')
revenue
```

### Необходимо сформировать  аналогичную таблицу по типам выручки с указанием адреса клиента

```python
revenue2 = client_base
revenue2 = revenue2.merge(air, on = 'client_id', how='left')
revenue2 = revenue2.merge(auto, on = 'client_id', how='left')
revenue2 = revenue2.merge(rzd, on = 'client_id', how='left')
revenue2
```
